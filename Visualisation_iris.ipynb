## Visualising Iris Dataset using Scatter Plots between Features

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
name = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_df = pd.read_csv("iris.data", names=name, sep=",")
```

```python
print(iris_df)
iris_df.dtypes
```

```python
sns.scatterplot(x='sepal_width', y='sepal_length', data=iris_df, hue='class', style='class', markers=["o", "s", "D"])
plt.title("Sepal Width Vs. Sepal Length")
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris_df, hue='class', style='class', markers=["o", "s", "D"])
plt.title("Sepal Length Vs. Petal Length")
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', data=iris_df, hue='class', style='class', markers=["o", "s", "D"])
plt.title("Sepal Length Vs. Petal Width")
plt.show()
```

```python
sns.scatterplot(x='sepal_width', y='petal_length', data=iris_df, hue='class', style='class', markers=["o", "s", "D"])
plt.title("Sepal Width Vs. Petal Length")
plt.show()
```

```python
sns.scatterplot(x='sepal_width', y='petal_width', data=iris_df, hue='class', style='class', markers=["o", "s", "D"])
plt.title("Sepal Length Vs. Petal Length")
plt.show()
```

```python
sns.scatterplot(x='petal_width', y='petal_length', data=iris_df, hue='class', style='class', markers=["o", "s", "D"])
plt.title("Petal Width Vs. Petal Length")
plt.show()
```

```python
sns.pairplot(data=iris_df, hue='class', markers=["o", "s", "D"], diag_kind="kde")
plt.title("Pair plot between features", loc="center")
plt.show()
```
