```python
img = cv2.imread("DATA/00-puppy.jpg")
```

```python
plt.imshow(img)
```

This is because the way matplotlib and opencv orders the rgb channels in different manner.

Matplotlib -> RGB
CV2 -> BGR

```python
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 2 RGB Convert
```

```python
plt.imshow(fixed_img)
```

```python
flipped_image = cv2.flip(fixed_img, -1)
```

```python
plt.imshow(flipped_image)
```
