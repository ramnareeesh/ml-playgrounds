```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("exercise_decision_trees.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
x0 = encoder.fit_transform(df.iloc[:, 0])
x1 = encoder.fit_transform(df.iloc[:, 1])
x2 = encoder.fit_transform(df.iloc[:, 2])
x3 = encoder.fit_transform(df.iloc[:, 3])
y = encoder.fit_transform(df.iloc[:, 4])
```

```python
X = np.vstack((x0, x1, x2, x3))
X
```

```python
X = X.T
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(criterion="entropy")
classifier.fit(X_train, y_train)
```

```python
plt.figure(figsize=(12,8))

from sklearn import tree

tree.plot_tree(classifier.fit(X, y))
plt.show()
```

```python
classifier.score(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```

```python
classifier.predict(X=[[2, 2, 0, 0]])
```
