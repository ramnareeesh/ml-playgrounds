```python
import numpy as np
```

```python
p_vals = [0.75, 0.25]  # purple p-value, white p-value
no_plants = 929
```

```python
simulations = 5000
simulated_freq = np.random.binomial(no_plants, p_vals[0], size=simulations)
```

```python
import matplotlib.pyplot as plt
plt.hist(simulated_freq)
plt.grid(True)
plt.xlabel("Statistic")
plt.ylabel("Frequency")
plt.scatter(705, 0, color='red')
plt.show()
```
