```python
import numpy
from numpy import random as rnd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population_dist = [0.26, 0.74]
```

```python
rnd.binomial(n=100, p=0.26)
```

```python
no_simulations = 1000
for i in range(no_simulations):
    black_chosen = rnd.binomial(100, 0.26, i)
```

```python
black_chosen
```

```python
no_blacks = pd.DataFrame(black_chosen, columns=["No of Black Chosen for Jury"])
```

```python
no_blacks
```

```python
plt.grid(True, linestyle='--', linewidth=0.5)
plt.xlabel(no_blacks.columns[0])
plt.ylabel("Units")
plt.hist(no_blacks)
plt.scatter(8, 0, color= "red")
plt.show()
```
