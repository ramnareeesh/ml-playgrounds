## Imports

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
breast_df = pd.read_csv("../data/breast-cancer.csv")
breast_df
```

```python
breast_df.dropna(axis=1, inplace=True)
breast_df
```

```python
breast_df["id"].value_counts().max()
```

```python
breast_df["diagnosis"].value_counts()
```

```python
X = breast_df.drop(["diagnosis", "id"], axis=1)
X
```

```python
y = breast_df["diagnosis"]
y
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y= le.fit_transform(y)
```

```python
y
```

M -> 1
B -> 0

## Dataset split for Test and Train

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=69)
```

```python
X_train
```

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logistic_regressor = LogisticRegression()
logistic_regressor.fit(X_train, y_train)
```

```python
lr_pred = logistic_regressor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
acc_lr = accuracy_score(y_test, lr_pred)
precision_lr = precision_score(y_test, lr_pred)
recall_lr = recall_score(y_test, lr_pred)
print("\nAccuracy: ", acc_lr)
print("Precision: ", precision_lr)
print("Recall: ", recall_lr, "\n")
```

```python
from sklearn.metrics import confusion_matrix
c_matrix_lr = confusion_matrix(y_test, lr_pred)
c_matrix_lr
```

### SVM Model

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)
```

```python
svm_pred = svm.predict(X_test)
```

```python
acc_svm = accuracy_score(y_test, svm_pred)
precision_svm = precision_score(y_test, svm_pred)
recall_svm = recall_score(y_test, svm_pred)
print("\nAccuracy: ", acc_svm)
print("Precision: ", precision_svm)
print("Recall: ", recall_svm, "\n")
```

```python
c_matrix_svm = confusion_matrix(y_test, svm_pred)
c_matrix_svm
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
dtc_pred = dtc.predict(X_test)
```

```python
acc_dtc = accuracy_score(y_test, dtc_pred)
precision_dtc = precision_score(y_test, dtc_pred)
recall_dtc = recall_score(y_test, dtc_pred)
print("\nAccuracy: ", acc_dtc)
print("Precision: ", precision_dtc)
print("Recall: ", recall_dtc, "\n")
```

```python
c_matrix_dtc = confusion_matrix(y_test, dtc_pred)
c_matrix_dtc
```

```python
from sklearn.metrics import roc_curve, auc
fpr_lr, tpr_lr, _ = roc_curve(y_test, lr_pred)
auc_lr = auc(fpr_lr, tpr_lr)
label_lr = "Logistic Regression - AUC: " + str(round(auc_lr, 2))

fpr_svm, tpr_svm, _ = roc_curve(y_test, svm_pred)
auc_svm = auc(fpr_svm, tpr_svm)
label_svm = "Support Vector Machine - AUC: " + str(round(auc_svm, 2))

fpr_dtc, tpr_dtc, _ = roc_curve(y_test, dtc_pred)
auc_dtc = auc(fpr_dtc, tpr_dtc)
label_dtc = "Decision Tree Classifier - AUC: " + str(round(auc_dtc, 2))

plt.plot(fpr_lr, tpr_lr, label=label_lr)
plt.plot(fpr_svm, tpr_svm, label=label_svm)
plt.plot(fpr_dtc, tpr_dtc, label=label_dtc)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROCs - all models")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Confusion Matrix - Logistic Regression")
sns.heatmap(c_matrix_lr, annot=True, cmap="Blues")
plt.show()
```
